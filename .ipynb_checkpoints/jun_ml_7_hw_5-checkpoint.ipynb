## Урок 8. Домашнее задание начального уровня


Обучите модель DBSCAN с параметром *min_samples=8*. Визуализируйте результат.

```python
from sklearn.cluster import DBSCAN
import numpy as np
import pickle
from matplotlib import pyplot as plt

%matplotlib inline

with open('clustering.pkl', 'rb') as f:
    data_cluster = pickle.load(f)

X, y = data_cluster['X'], data_cluster['Y']
db = DBSCAN(min_samples=8)
db.fit(X)


fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(X[:,0], X[:,1], c=db.labels_)
plt.show()


```
