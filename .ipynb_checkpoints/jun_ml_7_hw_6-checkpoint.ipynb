### 10.1 Домашнее задание начального уровня

Обучите *k-means* для количества кластеров $k=4$ на датасете `data/clustering.pkl`. Как изменилась метрика силуэтта? Стало лучше или хуже? Какое количество кластеров лучше, два или четыре?

```python
from sklearn.cluster import KMeans
import numpy as np
import pickle
from matplotlib import pyplot as plt
from sklearn.metrics import silhouette_score, adjusted_mutual_info_score

%matplotlib inline

with open('clustering.pkl', 'rb') as f:
    data_cluster = pickle.load(f)

X, y = data_cluster['X'], data_cluster['Y']

kmns2 = KMeans(n_clusters=2).fit(X)
s2 = silhouette_score(X, kmns2.labels_)
kmns4 = KMeans(n_clusters=4).fit(X)
s4 = silhouette_score(X, kmns4.labels_)

fig, axes = plt.subplots(2)
axes[0].scatter(X[:,0], X[:,1], c=kmns2.labels_)
axes[1].scatter(X[:,0], X[:,1], c=kmns4.labels_)
plt.show()

s2, s4 # Т.е. два кластера лучше

```


### 10.2 Домашнее задание начального уровня

Обучите *k-means* для количества кластеров $k=4$ на датасете `data/clustering.pkl`. Как изменилась метрика *AMI*? Стало лучше или хуже? Какое количество кластеров лучше по метрике AMI?

```python
ami2 = adjusted_mutual_info_score(y, kmns2.labels_)
ami4 = adjusted_mutual_info_score(y, kmns4.labels_)

ami2, ami4 # Т.е. два кластера тоже лучше
```

1. Каково качество кластеризации по AMI

```python
import pandas as pd

data = pd.read_csv('clustering_hw.csv')
X, y = data[['x1', 'x2']].values, data['y'].values

fig, ax = plt.subplots(figsize=(4, 3))
ax.scatter(X[:,0], X[:,1], c=y)
plt.show()

n = list(range(2, 10+1))
nums = []


for i in range(2, 10+1):
    kmns = KMeans(n_init=10, n_clusters=i).fit(X)
    nums.append(silhouette_score(X, kmns.labels_))

fig, ax = plt.subplots(figsize=(4, 3))
ax.plot(n, nums, marker='o')
plt.show()

# Похоже, что три класса это наилучший вариант

kmns = KMeans(n_init=10, n_clusters=3).fit(X)
kmns.labels_

# np.where(abs(X[:,0] - 5.0) < 0.01) 
np.where(abs(X[:,1] - 8.0) < 0.01)
# X[141]

np.where(abs(X[:,0] - 0.0) < 0.01) 
# np.where(abs(X[:,1] - 5.0) < 0.01) 
# X[288]

print(y[244], y[288], "      ", kmns.labels_[244], kmns.labels_[288])

# (5, 8) относится к классу 2, а (0, 5) -- к классу 1, т.е. к разным. 

adjusted_mutual_info_score(y, kmns.labels_)
```
